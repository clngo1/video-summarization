#Tier 3 Video: Pesto (Short film w/ complex movement and subtitles)

```python
#segment video into frames + upload frames Google Drive folder
```

```python
cam = cv2.VideoCapture("/content/drive/MyDrive/Video Summarizer Model/Tier 1 - Pesto Short Film.mp4")

try:

    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):

    # reading from frame
    ret,frame = cam.read()

    if ret:
        # if video is still left continue creating images
        #name = './data/frame' + str(currentframe) + '.jpg'
        name = '/content/drive/MyDrive/Video Summarizer Model/Tier 3 Frames/Tier 3' + str(currentframe) + '.jpg'
        print ('Creating...' + name)

        # writing the extracted images
        cv2.imwrite(name, frame)

        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```python
#captioning each frame from the tier 3 video
```
